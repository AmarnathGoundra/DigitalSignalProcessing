```python
import numpy as np
from scipy.fftpack import fft, fftfreq
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
fs = 300
t = np.linspace(0,1,fs)
#linespace divides range into 10 equal lines
x = np.sin(2*np.pi*5*t)
#x = np.sin(2*np.pi*50*t)
plt.subplot(221)
plt.plot(t,x,linewidth=3,color='k')
#linewidth changes the thickness of graph line
X=fft(x)
#coverts time domain values into Frequency Domain values
#fft stands for fast fourier transform
XPSD=np.abs(X)**2
#abs stands for modulus
XPSD/=len(x)**2
w=fftfreq(len(X),1/fs)
plt.subplot(222)
plt.plot(w,XPSD,linewidth=3,color='m')
plt.title('spectrum of noice free signal')

noise=np.random.normal(0,1,fs)

x=np.sin(2*np.pi*5*t)+noise
# x=np.sin(2*np.pi*50*t)+noise
plt.subplot(223)
plt.plot(t,x,linewidth=3,color='k')
plt.title('noisy signals')
X=fft(x)     
XPSD=np.abs(X)**2
XPSD/=len(X)**2
w=fftfreq(len(X),1/fs)
plt.subplot(224)
plt.plot(w,XPSD,linewidth=3,color='m')
plt.title('spectru, of noisy signal')

plt.savefig('fds on signals.jpg')
```
