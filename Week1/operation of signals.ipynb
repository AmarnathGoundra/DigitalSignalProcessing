```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(20,20))

t=np.linspace(-30,30,500)

x=5*((t>=-10) & (t<=10))
plt.subplot(441)
plt.plot(t,x)
plt.title("Original gate")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x2=5*(((t+4)>=-10) & ((t+4)<=10))
plt.subplot(443)
plt.plot(t,x2)
plt.title("Time shifting(Left)")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x3=5*(((t-4)>=-10) & ((t-4)<=10))
plt.subplot(444)
plt.plot(t,x3)
plt.title("Time shifting(Right)")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x4=5*(((2*t)>=-10) & ((t*2)<=10))
plt.subplot(445)
plt.plot(t,x4)
plt.title("Time scaling(Compress)")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x5=5*(((t/2)>=-10) & ((t/2)<=10))
plt.subplot(446)
plt.plot(t,x5)
plt.title("Time scaling(expand)")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x6=x/2
plt.subplot(447)
plt.plot(t,x6)
plt.title("Attenuation")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x7=x*4
plt.subplot(448)
plt.plot(t,x7)
plt.title("Amplification")
plt.xlabel("Time")
plt.ylabel("Amplitute")

t=-t
x1=x
plt.subplot(442)
plt.plot(t,x1)
plt.title("Reversal gate")
plt.xlabel("Time")
plt.ylabel("Amplitute")

plt.savefig("Operation on signals.jpg")
```
