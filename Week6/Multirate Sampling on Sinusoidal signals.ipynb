```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
t=np.linspace(0,1,100)
x=np.sin(2*np.pi*5*t)

plt.figure(figsize=(10,15))
plt.subplot(411)
plt.stem(x)
plt.title("Original signal")

y=signal.resample(x,int(len(x)*0.5))
plt.subplot(412)
plt.stem(y)
plt.title("Decimated signal")

y1=signal.resample(x,len(x)*3)
plt.subplot(413)
plt.stem(y1)
plt.title("Interpolated Signal")

y2=signal.resample(x,int(len(x)*1.5))
plt.subplot(414)
plt.stem(y2)
plt.title("Fractional sampling rate")

plt.savefig("Multirate Sampling.jpg")
```

```python
t=np.linspace(0,1,100)
x=np.sin(2*np.pi*5*t)
x1=np.sin(2*np.pi*10*t)
x2=np.sin(2*np.pi*3*t)

plt.figure(figsize=(10,15))

plt.subplot(311)
plt.plot(x)
plt.title("Original signal")

plt.subplot(312)
plt.plot(x1)
plt.title("Increased frequency signal")

plt.subplot(313)
plt.plot(x2)
plt.title("Decreased frequency signal")

plt.savefig("Change in freq & sampling rate of signal.jpg")
```
