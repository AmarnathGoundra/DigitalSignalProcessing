Generation of Signals

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,1,500)
x1=np.sin(2*np.pi*5*t)

fig=plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.plot(t,x1,linewidth=4)
plt.title("Sine wave")
plt.xlabel("Time")
plt.ylabel("Amplitute")

x2=np.exp(2*np.pi*5*t)
plt.subplot(2,2,2)
plt.plot(t,x2)
plt.title("exponential wave")
plt.xlabel("Time")
plt.ylabel("Amplitute")

t=np.linspace(-15,15,100)
x3=5*((t>=-10) & (t<=10))
plt.subplot(2,2,3)
plt.plot(t,x3,color="r")
plt.title("Gate wave")
plt.xlabel("Time")
plt.ylabel("Amplitute")

t=np.linspace(-1,1,1000)
x4=np.sinc(2*np.pi*5*t)
plt.subplot(224)
plt.plot(t,x4)
plt.title("sinc wave")
plt.xlabel("Time")
plt.ylabel("Amplitute")

plt.savefig("signal generation.jpg")
```
