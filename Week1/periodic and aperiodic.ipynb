```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(20,20))

t=np.linspace(-1,1,5000)

x=np.cos(2*np.pi*15*t)+np.cos(2*np.pi*10*t)
x1=np.cos(2*np.pi*15*t)+np.cos(2*10*t)
x2=np.cos(2*15*t)+np.cos(2*10*t)
x3=3*np.exp(-10*t)

plt.subplot(221)
plt.plot(t,x)
plt.title("Periodic signal")
plt.xlabel("Time")
plt.ylabel("Amplitute")

plt.subplot(222)
plt.plot(t,x1)
plt.title("Aperiodic signal")
plt.xlabel("Time")
plt.ylabel("Amplitute")

plt.subplot(223)
plt.plot(t,x2)
plt.title("Periodic signal")
plt.xlabel("Time")
plt.ylabel("Amplitute")

plt.subplot(224)
plt.plot(t,x3)
plt.title("Aperiodic signal")
plt.xlabel("Time")
plt.ylabel("Amplitute")

plt.savefig("periodic & aperiodic.jpg")
```
