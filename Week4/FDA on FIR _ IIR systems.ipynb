```python
import matplotlib.pyplot as plt
import numpy as np
from  matplotlib.pyplot import axvline,axhline
from scipy import signal
```

```python
plt.figure(figsize=(10,15))
plt.suptitle('FDA on FIR and IIR filters using inbuilt codes')

a=0.8;b=0.9

#FIR LPF
num=[1,a]
den=[1]
w,H=signal.freqz(num,den)

#Magnitude spectrum
plt.subplot(421)
plt.plot(w,abs(H))
plt.title('magnitude of FIR LPF in FDA')

#Pole zero plot
z,p,k=signal.tf2zpk(num,den)
t=np.linspace(0,1,100)
r=np.exp(1j*2*np.pi*t)

plt.subplot(422)
plt.plot(r.real,r.imag,'k')
plt.title('pole zero plot-FIR LPF')
axvline(0)
axhline(0)
plt.plot(p.real,p.imag,'xr',markersize=20,alpha=0.9)
plt.plot(z.real,z.imag,'or',markersize=20,alpha=0.9)

#FIR HPF
num=[1,-a]
den=[1]
w,H=signal.freqz(num,den)

#Magnitude spectrum
plt.subplot(423)
plt.plot(w,abs(H))
plt.title('magnitude of FIR HPF in FDA')

#Pole zero plot
z,p,k=signal.tf2zpk(num,den)
t=np.linspace(0,1,100)
r=np.exp(1j*2*np.pi*t)

plt.subplot(424)
plt.plot(r.real,r.imag,'k')
plt.title('pole zero plot-FIR HPF')
axvline(0)
axhline(0)
plt.plot(p.real,p.imag,'xr',markersize=20,alpha=0.9)
plt.plot(z.real,z.imag,'or',markersize=20,alpha=0.9)

#IIR LPF
num=[1]
den=[1,-b]
w,H=signal.freqz(num,den)

#Magnitude spectrum
plt.subplot(425)
plt.plot(w,abs(H))
plt.title('magnitude of IIR LPF in FDA')

#Pole zero plot
z,p,k=signal.tf2zpk(num,den)
t=np.linspace(0,1,100)
r=np.exp(1j*2*np.pi*t)

plt.subplot(426)
plt.plot(r.real,r.imag,'k')
plt.title('pole zero plot-IIR LPF')
axvline(0)
axhline(0)
plt.plot(p.real,p.imag,'xr',markersize=20,alpha=0.9)
plt.plot(z.real,z.imag,'or',markersize=20,alpha=0.9)

#IIR HPF
num=[1]
den=[1,b]
w,H=signal.freqz(num,den)

#Magnitude spectrum
plt.subplot(427)
plt.plot(w,abs(H))
plt.title('magnitude of HIR HPF in FDA')

#Pole zero plot
z,p,k=signal.tf2zpk(num,den)
t=np.linspace(0,1,100)
r=np.exp(1j*2*np.pi*t)

plt.subplot(428)
plt.plot(r.real,r.imag,'k')
plt.title('pole zero plot-IIR HPF')
axvline(0)
axhline(0)
plt.plot(p.real,p.imag,'xr',markersize=20,alpha=0.9)
plt.plot(z.real,z.imag,'or',markersize=20,alpha=0.9)

plt.savefig('FDA on FIR and IIR  systems using inbuilt codes.jpg')
```

```python
plt.figure(figsize=(12,15))
plt.suptitle('FDA on FIR and IIR systems using MATHEMATICAL expressions')

a=0.8
b=0.8

w1=np.linspace(0,np.pi,512)
m=len(w1)
H1=np.zeros((m,1),dtype="complex")

for i in range(m):
    H1[i]=1+a*np.exp(-1j*w1[i])
plt.subplot(421)
plt.plot(w1,abs(H1))
plt.title('magnitude of FIR LPF in FDA')


for i in range(m):
    H1[i]=1-a*np.exp(-1j*w1[i])
plt.subplot(422)
plt.plot(w1,abs(H1))
plt.title('magnitude of FIR HPF in FDA')


for i in range(m):
    H1[i]=1/(1-b*np.exp(-1j*w1[i]))
plt.subplot(423)
plt.plot(w1,abs(H1))
plt.title('magnitude of IIR LPF in FDA')


for i in range(m):
    H1[i]=1/(1+b*np.exp(-1j*w1[i]))
plt.subplot(424)
plt.plot(w1,abs(H1))
plt.title('magnitude of IIR HPF in FDA')

plt.savefig('FDA on FIR and IIR systems using MATHEMATICAL expressions.jpg')
```
