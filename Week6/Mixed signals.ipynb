```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import soundfile as sf
import sounddevice as sd

x1,fs1=sf.read('Voicesp.wav')

sd.play(x1,fs1)
sd.wait()

x2,fs2=sf.read('chirp12_4.8k.wav')

sd.play(x2,fs2)
sd.wait()

m1=len(x1)
m2=len(x2)

y4=np.append(x1,x2)

sd.play(y4,fs1)
sd.wait()

if(m1>m2):
    y1=np.append(x2,np.zeros(m1-m2))
    y=y1+x1
else:
    y2=np.append(x1,np.zeros(m2-m1))
    y=y2+x2
    
sd.play(y,fs1)
sd.wait()

sf.write('freq_speech1.wav',x1,fs1)
sf.write('freq_speech2.wav',x2,fs2)
sf.write('freq_concatanated_speech3.wav',y4,fs1)
sf.write('freq_mixed_speech4.wav',y,fs1)
        
plt.figure(figsize=(15,15))

plt.subplot(221)
plt.plot(x1)
plt.title('signal1')

plt.subplot(222)
plt.plot(x2)
plt.title('signal2')

plt.subplot(223)
plt.plot(y4)
plt.title('concatanated signal')

plt.subplot(224)
plt.plot(y)
plt.title('mixed signal')

plt.savefig('concatanation and mixing of 2 signals.jpg')
```
