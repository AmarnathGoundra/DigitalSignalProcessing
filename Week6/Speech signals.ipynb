```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import soundfile as sf
import sounddevice as sd
```

```python
x1,fs1=sf.read("swa_hap_1.wav")

sd.play(x1,fs1)
sd.wait()

sd.play(x1,fs1*0.5)
sd.wait()

sd.play(x1,fs1*3)
sd.wait()

sd.play(x1,fs1*1.5)
sd.wait()

sf.write("Freq_inc.wav",x1,int(fs1*0.5))
sf.write("Freq_dec.wav",x1,int(fs1*3))

y=signal.resample(x1,int(len(x1)*0.5))

y1=signal.resample(x1,int(len(x1)*3))

y2=signal.resample(x1,int(len(x1)*1.5))

sd.play(y,fs1*0.5)
sd.wait()

sd.play(y1,fs1*3)
sd.wait()

sd.play(y2,fs1*1.5)
sd.wait()

sf.write("upsampled_speech.wav",y,fs1)
sf.write("downsampled_speech.wav",y1,fs1)
sf.write("up&downsampled_speech.wav",y2,fs1)

plt.figure(figsize=(15,15))

plt.subplot(411)
plt.plot(x1)
plt.title("Original signal")

plt.subplot(412)
plt.plot(y)
plt.title("Decimated signal")

plt.subplot(413)
plt.plot(y1)
plt.title("Interpolated signal")

plt.subplot(414)
plt.plot(y2)
plt.title("fractional samplong signal")

plt.savefig("Speech signal.jpg")
```
