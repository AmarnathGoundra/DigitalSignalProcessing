# # NEGATIVE OF AN IMAGE

#Find an negative of an Image.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

plt.figure(figsize=(10,10))

pout=plt.imread('circles.png')
b1=1-pout

plt.subplot(1,2,1)
plt.imshow(pout)
plt.title('original image')

plt.subplot(1,2,2)
plt.imshow(b1)
plt.title('negative image')
plt.savefig('negative of given image1.jpg')
```

# # MULTIRATE SAMPLING

#Perform multirate sampling on 2D signals.

Changing Sampling Rate.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

plt.figure(figsize=(10,10))

img1=plt.imread('onion.png')
m,n,c=np.shape(img1)

y=signal.resample(img1,int(m*0.5))
y1=signal.resample(img1,m*3)
y2=signal.resample(img1,int(m*3/2))

plt.subplot(2,2,1)
plt.imshow(img1)
plt.title('original image')

plt.subplot(2,2,2)
plt.imshow(y)
plt.title('downsampled image')

plt.subplot(2,2,3)
plt.imshow(y1)
plt.title('upsampled image')

plt.subplot(2,2,4)
plt.imshow(y2)
plt.title('fractionally sampled image')

plt.savefig('multirate sampling of image.jpg')
```

Changing Frequency Rate

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

plt.figure(figsize=(10,10))

img1=plt.imread('onion.png')
m,n,c=np.shape(img1)

y=img1*0.5
y1=img1*3
y2=img1*3/2

plt.subplot(2,2,1)
plt.imshow(img1)
plt.title('original image')

plt.subplot(2,2,2)
plt.imshow(y)
plt.title('decreased frequency image')

plt.subplot(2,2,3)
plt.imshow(y1)
plt.title('increased frequency image')

plt.subplot(2,2,4)
plt.imshow(y2)
plt.title('fractionally increased frequency image')

plt.savefig('frequency change of image.jpg')
```

# # ADDITION OF IMAGES

#Mix/Add two images


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage.transform import resize


plt.figure(figsize=(20,20))

img1=plt.imread('peppers.png')
m,n,c=np.shape(img1)

img2=plt.imread('toysnoflash.png')

img4=resize(img2,[m,n])
img3=img1+img4

plt.subplot(3,1,1)
plt.imshow(img1)
plt.title('image1')

plt.subplot(3,1,2)
plt.imshow(img2)
plt.title('image2')

plt.subplot(3,1,3)
plt.imshow(img3)
plt.title('added image')

plt.savefig('added images.jpg')
```
