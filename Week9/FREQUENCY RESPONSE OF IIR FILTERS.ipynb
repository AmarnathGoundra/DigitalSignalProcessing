## BUTTERWORTH FILTER (1A, 1B)

IIR BUTTERWORTH FILTER

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

plt.figure(figsize=(10,5));

wc=0.6;
n=3;

num,den=signal.butter(n,wc);
w,H=signal.freqz(num,den);
w=w/(np.pi);

plt.subplot(1,2,1);
plt.plot(w,(20*np.log10(np.abs(H))),linewidth=3);
plt.ylim(-15,1);
plt.title('frequency response using butterworth')
plt.grid()

plt.savefig('frequency response using butterworth.jpg')
```

## Plot the frequency response of an IIR Butterworth filter for different cutoff frequency keeping the order of the filter fixed

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

plt.figure(figsize=(12,5));

wc=[0.4,0.5,0.6,0.7,0.8];
n=3;

for i in range(len(wc)):
    num,den=signal.butter(n,wc[i]);
    w,H=signal.freqz(num,den);
    w=w/(np.pi);
    plt.subplot(1,2,1);
    plt.plot(w,(20*np.log10(np.abs(H))),linewidth=3,label=wc[i]);
    plt.ylim(-15,1);
    plt.title('changing cut-off frequncy');
    plt.legend(loc='upper right');
    plt.grid();

plt.savefig('frequency response using butterworth and changing cut-off frequency.jpg')
```

## Plot the frequency response of an IIR Butterworth filter for different orders of the filter  keeping the cutoff frequency fixed
