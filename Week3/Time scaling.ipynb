```python
import numpy as np
from scipy.fftpack import fft, fftfreq
import matplotlib.pyplot as plt
```

```python
fs=250
t=np.linspace(-30,30,fs)
x1=(5)*((t>=-10)&(t<=10))
X1=fft(x1)
w=fftfreq(len(x1),1/fs)
plt.figure(figsize=(18,18))

plt.subplot(331)
plt.plot(t,x1,color='k')
plt.title('rectangular pulse $x1(t)$')

plt.subplot(332)
plt.plot(w,abs(X1),color='m')
plt.title('mag of rectangular pulse ')

plt.subplot(333)
plt.plot(w,np.angle(X1),color='m')
plt.title('phase of rectangular pulse ')

x2=(5)*((2*t>=-10)&(t*2<=10))
X2=fft(x2)

plt.subplot(334)
plt.plot(t,x2,color='k')
plt.title('compressed rectangular pulse $x2(t)$')

plt.subplot(335)
plt.plot(w,abs(X2),color='m')
plt.title('mag of compressed rectangular pulse')

plt.subplot(336)
plt.plot(w,np.angle(X2),color='m')
plt.title('phase of compressed rectangular pulse')


x3=(5)*((t/2>=-10)&(t/2<=10))
X3=fft(x3)
         
plt.subplot(337)
plt.plot(t,x3,color='k')
plt.title('expanded rectangular pulse $x3(t)$')

plt.subplot(338)
plt.plot(w,abs(X3),color='m')
plt.title('mag of expanded rectangular pulse')

plt.subplot(339)
plt.plot(w,np.angle(X3),color='m')
plt.title('phase of expanded rectangular pulse')

plt.savefig('time scaling.jpg')
```
