```python
import matplotlib.pyplot as plt
import numpy as np
from  matplotlib.pyplot import axvline,axhline
from scipy import signal
```

```python
r=0.9; theta=3*np.pi/8

a=[1] #Resonator or bandpass filter
b=[1,-2*r*np.cos(theta),r*r]
w,H= signal.freqz(a,b)

fig = plt.figure(figsize=(10,10))

fig.add_subplot(221)
plt.plot(w, np.abs(H))

#digital filter pole zero map
z,p,k = signal.tf2zpk(a,b)

fig.add_subplot(222)

t=np.linspace(0,1,100)
r=np.exp(1j*2*np.pi*t)
plt.plot(r.real, r.imag, 'k')

axvline(0)
axhline(0)

plt.plot(p.real, p.imag, 'xr', markersize=15, alpha=0.9)
plt.plot(z.real, z.imag, 'or', markersize=15, alpha=0.9)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import axvline, axhline
from scipy import signal

r=0.7; theta=np.pi/4; r1=1
a=[1,-2*r1*np.cos(theta),r1]  ## Notch filter or bandstop filter
b=[1,-2*r*np.cos(theta),r*r]
w, H = signal.freqz(a,b)

fig = plt.figure(figsize=(10,10))

fig.add_subplot(221)
plt.plot(w, np.abs(H))

# digital filter pole zero map
z,p,k = signal.tf2zpk(a,b)

fig.add_subplot(222)

t = np.linspace(0, 1, 100)
r = np.exp(1j*2*np.pi*t)
plt.plot(r.real,r.imag, 'k')

axvline(0) 
axhline(0) 
plt.plot(p.real, p.imag, 'xr', markersize=15, alpha=0.9)
plt.plot(z.real, z.imag, 'ob', markersize=15, alpha=0.9)
plt.grid()
plt.show()

```
