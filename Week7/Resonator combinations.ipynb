```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# for different radii
r = [.75,0.8, 0.85,0.9,0.95]
theta=np.pi/4
c = ['r','b','g','m','k']
a=[1]

plt.figure()

for i in range(len(r)):
    b=[1,-2*r[i]*np.cos(theta),r[i]*r[i]]
    w, H = signal.freqz(a,b)
    plt.plot(w, np.abs(H),color=c[i],linewidth=2,label=r[i])
    plt.legend(loc="upper right")

plt.savefig('resonator_radius_combinations.jpg')
# for different angles

theta=[0,np.pi/4,np.pi/2,3*np.pi/4,np.pi]
r=0.8
plt.figure()

for i in range(len(theta)):
    t=theta[i]
    b=[1,-2*r*np.cos(t),r*r]
    w, H = signal.freqz(a,b)
    plt.plot(w, np.abs(H),color=c[i],linewidth=2,label=theta[i])
    plt.legend(loc="upper right")
    
plt.savefig('resonator_theta_combinations.jpg')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# for different radii
r = [.75,0.8, 0.85,0.9,0.95]
theta = np.pi/4
c = ['r','b','g','m','k']
a = [1,-2*np.cos(theta),1]

plt.figure()

for i in range(len(r)):    
    b=[1,-2*r[i]*np.cos(theta),r[i]*r[i]]
    w, H = signal.freqz(a,b)
    plt.plot(w, np.abs(H),color=c[i],linewidth=2,label=r[i])
    plt.legend(loc="upper right")
    
plt.savefig('resonator_combinations_radius.jpg')

# for different angles

theta=[0,np.pi/4,np.pi/2,3*np.pi/4,np.pi]
r=0.8
plt.figure()

for i in range(len(theta)):
    t=theta[i]
    b=[1,-2*r*np.cos(t),r*r]
    w, H = signal.freqz(a,b)
    plt.plot(w, np.abs(H),color=c[i],linewidth=2,label=theta[i])
    plt.legend(loc="upper right")
    
plt.savefig('resonator_combinations_angle.jpg')
```
