```python
import matplotlib.pyplot as plt
import numpy as np
from  matplotlib.pyplot import axvline,axhline
from scipy import signal
```

```python
plt.figure(figsize=(10,15))
plt.suptitle('FDA on FIR and IIR filters using inbuilt codes')

#FIR LPF
num=[0.8,1.2]
den=[1,-0.6,-0.7]
w,H=signal.freqz(num,den)

#Magnitude spectrum
plt.subplot(421)
plt.plot(w,abs(H))
plt.title('magnitude of FIR LPF in FDA')

#Pole zero plot
z,p,k=signal.tf2zpk(num,den)
t=np.linspace(0,1,100)
r=np.exp(1j*2*np.pi*t)

plt.subplot(422)
plt.plot(r.real,r.imag,'k')
plt.title('pole zero plot-FIR LPF')
axvline(0)
axhline(0)
plt.plot(p.real,p.imag,'xr',markersize=20,alpha=0.9)
plt.plot(z.real,z.imag,'or',markersize=20,alpha=0.9)

plt.savefig('FDA on differentiation.jpg')
```

```python
plt.figure(figsize=(12,15))
plt.suptitle('FDA on FIR and IIR systems using MATHEMATICAL expressions')

w1=np.linspace(0,np.pi,1024)
m=len(w1)
H1=np.zeros((m,1),dtype="complex")

for i in range(m):
    H1[i]=(0.8+1.2*np.exp(-1j*w1[i]))/(1-0.6*np.exp(-1j*w1[i])-0.7*np.exp(-1j*w1[i]))
plt.subplot(211)
plt.plot(w1,abs(H1))
plt.title('magnitude of FIR LPF in FDA')

plt.savefig('FDA on differentiation using MATHEMATICAL expressions.jpg')
```
